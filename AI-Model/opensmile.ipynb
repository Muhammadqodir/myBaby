```python
import opensmile
import os
import pandas
import numpy
```

```python
smile = opensmile.Smile(
    feature_set=opensmile.FeatureSet.eGeMAPSv01a,
    feature_level=opensmile.FeatureLevel.LowLevelDescriptors,
)
```

```python
processed = smile.process_file('files/03-02-02-02-01-02-17.wav')
```

```python
processed
```

```python
files = os.listdir("./files")
files = ["files/"+s for s in files]

sorted_files = []
for file in files:
    if file[-4:] == ".wav":
        sorted_files.append(file)
```

```python
len(sorted_files)
```

```python
dataSet = smile.process_files(sorted_files)
```

```python
dataSet
```

```python
from sklearn.preprocessing import Normalizer
norm = Normalizer().fit(dataSet)
normalized_X = norm.transform(dataSet)
print("Before:")
print(dataSet[:1].values)
print("After:")
print(normalized_X[:1])
```

```python
def getEmotion(fileName):
    emotions = {
        "01":"neutral",
        "02":"calm",
        "03":"happy",
        "04":"sad",
        "05":"angry",
        "06":"fearful",
        "07":"disgust",
        "08":"surprised"
    }
    fileName = fileName.replace("files/", "")
    fileName = fileName.replace(".wav", "")
    ids = fileName.split("-")
    return emotions[ids[2]]

y_train = numpy.array([])
for file in sorted_files:
    y_train = numpy.append(y_train, getEmotion(file))
```

```python
y_train
```

```python
y_train_df = pandas.DataFrame(data=y_train, columns=["emotion"])
y_train_df
```

```python
Y = pandas.get_dummies(y_train_df["emotion"], prefix="emotion", dummy_na=False)
Y
```

```python
X = dataSet
```

```python
print(X.shape)
print(Y.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.05, random_state = 0)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
model_tree = DecisionTreeClassifier(max_depth=40)
model_tree.fit(X_train, y_train)
```

```python
y_predict = model_tree.predict(X_test)
y_predict[:10]
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, y_predict))
```

```python
from sklearn.model_selection import cross_validate
print("Cross validation:")
cv_res = cross_validate(model_tree, X, Y, cv=5)

print(cv_res["test_score"])
```
